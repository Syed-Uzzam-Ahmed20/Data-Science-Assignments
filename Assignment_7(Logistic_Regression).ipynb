```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 1. Data Exploration:
# a. Load the dataset and perform exploratory data analysis (EDA)
titanic_train = pd.read_csv('Titanic_train.csv')
titanic_train 
```

```python
titanic_test = pd.read_csv('Titanic_test.csv')
titanic_test
```

```python
# Combine train and test datasets for EDA
df = pd.concat([titanic_train, titanic_test], ignore_index=True)
df
```

```python
# b. Examine the features, their types, and summary statistics
print(df.info())                   # Check data types and missing values
print(df.describe())               # Summary statistics for numerical features
print(df.describe(include=['O']))  # Summary statistics for categorical features
```

```python
# c. Create visualizations
# Histograms
df.hist(figsize=(12, 8))
plt.title('Histograms of Numerical Features', fontsize=16)
plt.show()

# Boxplots
df.boxplot(figsize=(12, 8))
plt.title('Boxplots of Numerical Features', fontsize=16)
plt.show()

# Pair plots
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']])
plt.suptitle('Pair Plot of Selected Features', fontsize=16)
plt.show()
```

```python
# 2. Data Preprocessing:
#a. Handle missing values
import pandas as pd

# Impute missing values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
# Encode categorical variables:
# Label encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
X_cate = df[['Survived','Name','Sex','Ticket','Cabin','Embarked']]

X_cate['Survived']=LE.fit_transform(X_cate['Survived'])
X_cate['Name']=LE.fit_transform(X_cate['Name'])
X_cate['Sex']=LE.fit_transform(X_cate['Sex'])
X_cate['Ticket']=LE.fit_transform(X_cate['Ticket'])
X_cate['Cabin']=LE.fit_transform(X_cate['Cabin'])
X_cate['Embarked']=LE.fit_transform(X_cate['Embarked'])

X_cate.head() 
```

```python
# Standard scaling
X_cont = titanic_train[['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]

#Standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_cont_SS = SS.fit_transform(X_cont)
X_cont_SS = pd.DataFrame(X_cont_SS)
X_cont_SS.columns = list(X_cont)
X_cont_SS.head() 
```

```python
# Prepare the features and target variable
X = pd.concat([X_cont_SS, X_cate], axis=1)
print(X.shape)

y = X['Survived']
print(y.shape)
```

```python
from sklearn.impute import SimpleImputer

# Define the imputer
imputer = SimpleImputer(strategy='mean') 
X_imputed = imputer.fit_transform(X) 
X = pd.DataFrame(X_imputed, columns=X.columns) 
X  
```

```python
# 3. Model Building:
# a. Build a logistic regression model using appropriate libraries (e.g., scikit-learn).                                                                                  b. Train the model using the training data.

import pandas as pd
from sklearn.linear_model import LogisticRegression

# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Create a logistic regression model
model = LogisticRegression(multi_class='ovr', max_iter=200)

# b. Train the model using training data
model.fit(X_train, y_train) 
```

```python
# 4. Model Evaluation:
# a. Evaluate the performance of the model on the testing data 
y_pred = model.predict(X_test)
y_pred = pd.DataFrame(y_pred)
print(y_pred.value_counts())

# Get probability estimates for each class
y_proba = model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize

ac= accuracy_score(y_test, y_pred)
print("Accuracy score:", ac.round(2))

rs = recall_score(y_test, y_pred, average='weighted')
print("recall score:", rs.round(2))  

ps = precision_score(y_test, y_pred, average='weighted')
print("precision score:", ps.round(2)) 

f1 = f1_score(y_test, y_pred, average='weighted')
print("f1 score:", f1.round(2)) 

roc_auc = roc_auc_score(y_test, y_proba, average='weighted', multi_class='ovr')
print("roc-auc score:", roc_auc.round(2))  
```

```python
# Visualize the ROC curve:
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Binarize the output
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_bin.shape[1]

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves for each class
plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label='ROC curve of class {0} (area = {1:0.2f})'
                                   ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show() 
```

```python
# 5. Interpretation:
# a. Interpret the coefficients of the logistic regression model.
coefficients = model.coef_[0]

# Print coefficients with feature names
for feature, coef in zip(X.columns, coefficients):
    print(f"{feature}: {coef:.4f}")
```

```python
# Discuss the significance of features in predicting the target variable (survival probability in this case).
import pandas as pd

# Create a DataFrame for better visualization
col_importance = pd.DataFrame({'Feature': X.columns, 'Coefficient': coefficients})
col_importance = col_importance.sort_values('Coefficient', key=abs, ascending=False)
print(col_importance) 
```

```python
#Interview Questions:
# 1. What is the difference between precision and recall?
# Answer: Key Differences
#          1.Focus of Measurement:
#              Precision focuses on the accuracy of positive predictions made by the model.
#              Recall focuses on the model's ability to find all relevant positive instances in the dataset.

#          2.Error Type Penalized:
#                Precision penalizes false positives (type I errors).
#                Recall penalizes false negatives (type II errors).

#          3.Application Priorities:
#                 High Precision Priority: When the cost of a false positive is high.
#                 High Recall Priority: When the cost of a false negative is high.
```
